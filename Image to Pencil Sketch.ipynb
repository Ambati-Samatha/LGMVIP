# importing libraries

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def pencil_sketch(image_path):
    # Read the image in RGB format
    image = cv2.imread("cartoon.jpg")
    
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Invert the grayscale image
    inverted_image = cv2.bitwise_not(gray_image)
    
    # Apply Gaussian blur to the inverted image
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
    
    # Divide the grayscale image by the inverted blurry image
    pencil_sketch = cv2.divide(gray_image, blurred_image, scale=256.0)
    
    # Display the original image and the pencil sketch
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axs[0].set_title("Original Image")
    axs[0].axis("off")
    axs[1].imshow(pencil_sketch, cmap="gray")
    axs[1].set_title("Pencil Sketch")
    axs[1].axis("off")
    plt.show()
```

```python
# Providing the path to your image file
image_path = "sketch.jpg"
```

```python
# Calling the function to generate the pencil sketch
pencil_sketch(image_path)
```
